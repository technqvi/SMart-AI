```python
from google.cloud import bigquery
from google.cloud import storage

#https://pypi.org/project/gcsfs/
```

```python
client_bq = bigquery.Client()

sql = """

 SELECT  company,open_datetime, close_datetime,updated_at,severity_name ,brand,model,product_type
 FROM `pongthorn.SMartDW.incident` 
 WHERE DATE(open_datetime) > "2023-02-01" order by updated_at desc LIMIT 10
"""

query_result=client_bq.query(sql)
df=query_result.to_dataframe()
df.info()
df
```

```python
gcs_path="gs://pongthorn-data-exploration"
train_csv_file='incident_ml_train.csv'
df.to_csv(f"{gcs_path}/{train_csv_file}",index=False) 
```
